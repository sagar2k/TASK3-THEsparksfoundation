## For a given iris dataset create a Decision Tree classifier and visulaize it graphically(Prediction using Decision Tree Algorithm)



```python
#IMPORTING LIBRARIES
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sns
```

```python
#IMPORTING DATASET
iris=pd.read_csv('D:\iris.csv')
iris.head()
```

```python
#DROPPING UNNCESSARY COLUMNS
iris.drop('Id',axis='columns',inplace=True)
iris.head()
```

```python
features=iris[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y=iris['Species']
print(features.head(),'\n')
print(y.head())
```

```python
#Training the model
x_train,x_test,y_train,y_test=train_test_split(features,y,random_state=0)
dtc=DecisionTreeClassifier().fit(x_train,y_train)
print('Test accuracy {}, Train accuracy {}'.format(dtc.score(x_test,y_test),dtc.score(x_train,y_train)))
```

```python
dtc.predict(x_test)
```

```python
#EVALUATION METRICS
print(classification_report(y_test,dtc.predict(x_test)))
sns.heatmap(confusion_matrix(y_test,dtc.predict(x_test)),annot=True,cbar=False,
            xticklabels=y.unique(),yticklabels=y.unique());
plt.xlabel('Predicted Label');
plt.ylabel('True Label');
```

```python
plt.subplots(figsize=(18,18))
tree.plot_tree(dtc,feature_names=features.columns,class_names=y.unique(),filled=True);
```
